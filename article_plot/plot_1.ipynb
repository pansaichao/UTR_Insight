```python
def seq_info(seq_in):
    seq_len = []
    seq_GC_prec = []
    seq_GC_ratio = []
    seq_AT_ratio = []

    for i in range(len(seq_in)):
        i_len = len(seq_in[i])
        i_A_count = seq_in[i].count("A")
        i_T_count = seq_in[i].count("T")
        i_C_count = seq_in[i].count("C")
        i_G_count = seq_in[i].count("G")
        seq_len.append(i_len)

        if i_C_count != 0:
            seq_GC_ratio.append(i_G_count / i_C_count)
        else:
            seq_GC_ratio.append("na")
        if i_T_count != 0:
            seq_AT_ratio.append(i_A_count / i_T_count)
        else:
            seq_AT_ratio.append("na")
        seq_GC_prec.append((i_C_count + i_G_count) / i_len)

    return seq_len, seq_GC_prec, seq_GC_ratio, seq_AT_ratio
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 读取 CSV 文件
df = pd.read_csv('D:/code_work/article_plot/result_1/model_compare.csv')

# 设置绘图样式
sns.set(style="whitegrid")

# 设置字体为 Arial，并设置字体大小为 8 磅
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 8

# 获取不同的 Dataset 值
datasets = df['Dataset'].unique()

# 遍历 Dataset 值，分别绘制图形并保存为 svg 格式
for i, dataset in enumerate(datasets):
    # 根据当前的 Dataset 筛选数据
    df_subset = df[df['Dataset'] == dataset]
    
    # 调整图的大小，单位为英寸，下面提供三种不同版式的尺寸设置
    # 选择单栏图
    fig_width = 8.5 / 2.54  # 单栏图 8.5 cm = 3.35 英寸
    # 选择 1.5 栏图
    # fig_width = 12.0 / 2.54  # 1.5 栏图 12 cm = 4.72 英寸
    # 选择双栏图
    # fig_width = 18.0 / 2.54  # 双栏图 18 cm = 7.09 英寸

    # 设置图的高度，SCI投稿官网没有具体要求，这里设定为合适的比例
    fig_height = fig_width * 0.75  # 高度为宽度的0.75倍
    plt.figure(figsize=(fig_width, fig_height))

    # 创建柱状图
    ax = sns.barplot(data=df_subset, x="Length", y="Spearman R", hue="Model", palette='deep')

    # 设置图表标题、标签
    plt.title(f'{dataset}', fontsize=8)
    plt.xlabel("Length", fontsize=8)
    plt.ylabel("Spearman R", fontsize=8)

    # 设置y轴的刻度范围
    plt.ylim(0, 1.0)

    # 设置x轴和y轴刻度的字体大小，并旋转x轴标签防止重叠
    plt.xticks(fontsize=8, rotation=0)
    plt.yticks(fontsize=8)

    # 调整图例：将图例放在图外，移除图例标题，并设置字体大小
    plt.legend(title='', frameon=False, bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)

    # 设置轴线宽度
    plt.gca().spines['top'].set_linewidth(0.5)
    plt.gca().spines['right'].set_linewidth(0.5)
    plt.gca().spines['bottom'].set_linewidth(0.5)
    plt.gca().spines['left'].set_linewidth(0.5)

    # 隐藏右侧和上方的轴线
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    # 控制图例显示：第一个图不带图例，第二个图及以后带图例
    if i == 0:
        # 第一个图不显示图例
        ax.get_legend().remove()

    # 保存为 SVG 格式
    plt.savefig(f'D:/code_work/article_plot/result_1/{dataset}_barplot.svg', format='svg', bbox_inches='tight')

    # 显示图形
    plt.show()
```
